# Vector RAG Query

Use this notebook to call `query_vector_rag` with your own parameters.


```python
import os
import sys
from pathlib import Path

PROJECT_ROOT = Path("..").resolve()
if str(PROJECT_ROOT) not in sys.path:
    sys.path.append(str(PROJECT_ROOT))

from RAG.VectorRAG import query_vector_rag
```

```python
# Update these values to match your Neo4j vector index configuration.
question = "What insights can we surface about Graph RAG?"

vector_index_name = os.getenv("NEO4J_VECTOR_INDEX", "your_vector_index")
vector_node_label = os.getenv("NEO4J_VECTOR_NODE_LABEL", "YourNodeLabel")
vector_source_property = os.getenv("NEO4J_VECTOR_SOURCE_PROPERTY", "source_property")
vector_embedding_property = os.getenv("NEO4J_VECTOR_EMBEDDING_PROPERTY", "embedding_property")

# Set to None to rely on the default from the environment or VectorRAG module.
embedding_model = os.getenv("OPENAI_EMBED_MODEL") or None
```

```python
if vector_index_name == "your_vector_index":
    raise ValueError("Set vector_index_name to a valid index before running the query.")

response = query_vector_rag(
    question=question,
    vector_index_name=vector_index_name,
    vector_node_label=vector_node_label,
    vector_source_property=vector_source_property,
    vector_embedding_property=vector_embedding_property,
    embedding_model=embedding_model,
)

print(response)
```
